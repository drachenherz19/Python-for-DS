```python
#Advanced line plot:
#Defining the dataset:
a = np.arange (0,10,1)
b = 2*a + 5
#Changing the figure:
fig = plt.figure(figsize = (10,5))
#PLotting the datapoints:
plt.plot(a,b, linewidth = 2.0, linestyle ='-.', color = 'r', alpha = 0.5, marker = 'o')
plt.title("Advanced Line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['Line1'], loc = 'best')
plt.grid(True)
plt.show()
```

```python
#Multiple graphs in one single visualization:
x3 = np.arange(0,10,1)
y3 = 2*x3+5
y4 = 5*x3 +3
plt.subplot(1,2,1)
plt.plot(x3,y3)
plt.title("First Graph")
plt.subplot(1,2,2)
plt.plot(x3,y4)
plt.title("Second Graph")
plt.show()
```
