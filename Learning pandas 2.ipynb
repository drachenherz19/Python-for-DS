```python
import numpy as np
import pandas as pd
numpyarray = np.array(([5000,6000], ['John', 'James']))
df = pd.DataFrame({'name':numpyarray[1], 'salary':numpyarray[0]})
df
```

```python
#Performing merge operation between two dataframes
player = ['player1','player2', 'player 3']
points = [10, 9, 8]
title = ['game1', 'game2', 'game3']
df1= pd.DataFrame({'Player':player, 'Points': points,'Title': title})
df1
```

```python
player = ['player1', 'player5', 'player6']
power = ['punch', 'kick', 'elbow']
title = ['game1', 'game5', 'game6']
df2= pd.DataFrame({'Player':player, 'Power':power, 'Title':title})
df2
```

```python
#The player and title column are the same, so we can perform the merge operation on them
df1.merge(df2, on = 'Player', how = 'inner')
```

```python
#Left merge
df1.merge(df2, on='Player', how = 'left')
```

```python
#right merge
df1.merge(df2, on ='Player', how = 'right')
```

```python
#outer merge
df1.merge(df2, on='Player', how = 'outer')
```

```python
#Join is used when both tables have completely different values. Index values are used to join the values between the tables
player = ['player1', 'player5', 'player6']
power = ['punch', 'kick', 'elbow']
title = ['game1', 'game5', 'game6']
df4= pd.DataFrame({'Player':player, 'Power':power, 'Title':title}, index = ['L2','L3', 'L4'])
df4
```

```python
player = ['player1','player2', 'player 3']
points = [10, 9, 8]
title = ['game1', 'game2', 'game3']
df3= pd.DataFrame({'Player':player, 'Points': points,'Title': title}, index =['L1', 'L2', 'L3'])
df3
```

```python
#inner join
df3.join(df4, how='inner', lsuffix='left', rsuffix='right')
```

```python
#left join
df3.join(df4, how = 'left', lsuffix ='left', rsuffix='right')
```

```python
#right join
df3.join(df4, how='right', lsuffix='left', rsuffix='right')
```

```python
#outer join
df3.join(df4, how='outer', lsuffix='left', rsuffix='right')
```

```python
#Concatenation using Pandas
pd.concat([df3,df4])
```
