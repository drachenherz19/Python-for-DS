```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = {'Bananas':20, 'Mangoes':15, 'Cranberries':40, 'Peaches':25}
names = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10,5))
#Creating the bar plot
plt.bar(names, values)
plt.title("Bar Graph")
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```

```python
#Scatter Plot:
a = np.arange(10,100, 10)
b = np.arange(1,10,1)
x = [2,1,2,3,4,5,2,3,1]
plt.scatter(a,b, c='g', s=100, edgecolors='y', marker='o', alpha=0.5)
plt.scatter(a,x, c='black', s=150, edgecolors='o', marker='4', alpha = 1)
plt.legend(['b','x'], loc='best')
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.savefig('Scatterplot.png')
plt.show()
```

```python
#Histogram:
num = np.arange(10,100, 3)
print(num)
plt.title("Histogram")
plt.xlabel('Range of values')
plt.ylabel('Range of frequencies')
plt.hist(num, bins = [0,20,40,60,80,100], color='r', edgecolor = '#000000')
```

```python
#Box plot:
b1 = [10,20,12,3,11,19,72,15,40,60,23]
b2 = [34,31,35,37,38,57,34,100,12,40,41]
b3 = [20,22,24,25,26,21,27,29,30,100,19]
bdata = list([b1,b2,b3])
plt.boxplot(bdata, showmeans=True)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

```python
#Violin Plot:
b1 = [10,20,12,3,11,19,72,15,40,60,23]
b2 = [34,31,35,37,38,57,34,100,12,40,41]
b3 = [20,22,24,25,26,21,27,29,30,100,19]
vdata=list([b1,b2,b3])
plt.violinplot(vdata, showmeans=True, showmedians=True)
plt.grid(True)
plt.title("Violin plot")
plt.show()


```

```python
#Pie Chart:
label = ['Dog', 'Elephant', 'Cat', 'Bat', 'Gazelle']
sizes = [100,5,90,50,80]
plt.pie(sizes, labels = label)
plt.title('Pie Chart Demo')
plt.show()
```

```python
#Donut chart:
group_names = ['A', 'B', 'C']
group_size = [20,30,40]
size_centre = [5]
outer_ring = plt.pie(group_size, labels = group_names, radius = 2)
inner_white = plt.pie(size_centre, radius = 1, colors='w')
plt.show()
```
